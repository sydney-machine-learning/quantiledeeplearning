```python
import pandas as pd 
import matplotlib.pyplot as plt 
import seaborn as sns
```

Uni BD

```python
actual_data = pd.read_csv('bitcoin_uni_bdlstm_classic_actual.csv', index_col=0)  # Replace with your actual file path
pred_data = pd.read_csv('bitcoin_uni_bdlstm_classic_pred.csv', index_col=0)  # Replace with your actual file path
quantile_data = pd.read_csv('bitcoin_uni_bdlstm_quantile_results.csv',index_col=0)


# Select the specific row to visualize
row_index = 27 # change this to select different rows
actual_row = actual_data.iloc[row_index]
pred_row = pred_data.iloc[row_index]

# Reshape quantile data
quantiles = []
for i in range(5):
    quantiles.append(quantile_data.iloc[row_index, i*5:(i+1)*5])

# Time steps
time_steps = ['1', '2', '3', '4', '5']

# Set ggplot style and Helvetica font
sns.set_style('darkgrid')
plt.rcParams['font.family'] = 'Helvetica'

# Adjust the figure size
fig_width, fig_height = 8, 6  # Adjust these values to your preferred dimensions
quantile_label = [0.05, 0.25, 0.5, 0.75, 0.95]
# Plot actual and predicted values
plt.figure(figsize=(fig_width, fig_height))
plt.plot(time_steps, actual_row, marker='o', label='Actual', color='black')
plt.plot(time_steps, pred_row, marker='o', label='Predicted', color='brown', linestyle='--', linewidth=2)

# Plot quantiles
colors = ['#FFD700', '#FFA500', '#FF4500', '#DC143C', '#8B0000'] # Different shades for quantiles
for i in range(4):
    plt.fill_between(time_steps, quantiles[i], quantiles[i+1], color=colors[i], alpha=0.3, label=f'Quantile {quantile_label[i]} - {quantile_label[i+1]}')

plt.plot(time_steps, quantiles[2], label='Median', color=colors[2], linewidth=4, linestyle='--')


# Add labels and legend
plt.xlabel('Time Steps')
plt.ylabel('Values')
plt.legend(loc='upper left')
plt.grid(True)
plt.tight_layout()
plt.savefig('bitcoin_uni_bdlstm_quantile.png')
plt.show()
```
