```python
# import libraries - model
import torch
import torch.nn as nn
import torch.nn.functional as F
```

```python
# Arpit Kapoor implementation
def tilted_loss(q, y, f):
    e = (y-f)
    return torch.mean(torch.max(q*e, (q-1)*e), dim=-1)
```

```python
# q = quantile level
# y = true value
# f = predicted value

# e = residual (true - pred)
```

```python
# Alternative implementation - better interpretation
def tilted_loss(quantile, true_val, pred_val):
    residual = true_val - pred_val
    return torch.mean(torch.max(quantile*residual, (quantile - 1)*residual), dim = -1)
```

```python
class MLP(nn.Module):
    def __init__(self, input_size, hidden_size, output_size):
        super(MLP, self).__init__()
        self.fc1 = nn.Linear(input_size, hidden_size)
        self.fc2 = nn.Linear(hidden_size, output_size)

    def forward(self, x):
        x = F.relu(self.fc1(x))
        x = self.fc2(x)
        return x
```
